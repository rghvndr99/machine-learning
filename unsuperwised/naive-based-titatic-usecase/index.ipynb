```python
import pandas as pd
```

```python
pf=pd.read_csv("titanic.csv")
pf.head()
```

```python
pf=pf.drop(["PassengerId","Name","SibSp","Cabin","Embarked","Parch","Ticket"],axis="columns")
```

```python
pf.head()
```

```python
pc=pf.drop(["Ticket"],axis="columns")
pc.head()
```

```python
target= pc.Survived
inputs=pc.drop("Survived", axis="columns")
inputs.head()
```

```python
dummies= pd.get_dummies(pc.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis="columns")
inputs.head()
```

```python
f_input=inputs.drop(["Sex"], axis="columns")
```

```python
f_input.head()
```

```python
f_input.columns[f_input.isna().any()]
```

```python
f_input.Age=f_input.Age.fillna(f_input.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(f_input,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
cf=GaussianNB()
cf.fit(X_train,y_train)
```

```python
cf.score(X_test,y_test)
```

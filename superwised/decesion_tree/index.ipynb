```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
df=pd.read_csv("salaries.csv")
df
```

```python
inpt= df.drop("salary_more_then_100k",axis="columns")
output=df["salary_more_then_100k"]
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inpt['n_company']=le_company.fit_transform(inpt["company"])
inpt['n_job']=le_job.fit_transform(inpt["job"])
inpt['n_degree']=le_degree.fit_transform(inpt["degree"])
input_n=inpt.drop(["company","job","degree"],axis="columns")
input_n
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree=DecisionTreeClassifier()
cf=dec_tree.fit(input_n,output)
```

```python
cf.score(input_n,output)
```

```python
cf.predict([[3,0,0]])
```

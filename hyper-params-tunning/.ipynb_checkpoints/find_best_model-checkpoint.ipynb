```python
from sklearn import svm,datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

iris= datasets.load_iris()
```

```python
import pandas as pd
pf=pd.DataFrame(iris.data,columns=iris.feature_names)
pf.head()
```

```python
pf["flowers"]=iris.target
pf["flowers"]=pf["flowers"].apply(lambda x: iris.target_names[x])
pf[45:52]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size= 0.3)
```

```python
modelDisc= {
    'svm': {
        'model': svm.SVC(gamma="auto"),
        'params': {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        } 
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1,5,10],
        } 
    },
   'logistic_reg': {
        'model':  LogisticRegression(),
        'params': {
            'C': [1,5,10],
        } 
    },
}
```

```python
scores=[]
from sklearn.model_selection import GridSearchCV
for key,value in modelDisc.items():
    clf=GridSearchCV(value['model'],value['params'], cv=5, return_train_score=False)
    clf.fit(iris.data,iris.target)
    scores.append({
        'model': key,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

scores
```

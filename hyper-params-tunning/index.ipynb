```python
from sklearn import svm,datasets
iris= datasets.load_iris()
```

```python
iris
```

```python
import pandas as pd
pf=pd.DataFrame(iris.data,columns=iris.feature_names)
pf.head()
```

```python
pf["flowers"]=iris.target
pf["flowers"]=pf["flowers"].apply(lambda x: iris.target_names[x])
pf[45:52]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size= 0.3)
```

```python
model=svm.SVC(kernel='rbf',
    C=30,
    gamma='auto')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=30,gamma='auto'), iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='linear',C=30,gamma='auto'), iris.data,iris.target,cv=5)
```

```python
from sklearn.model_selection import GridSearchCV
cf=GridSearchCV(svm.SVC(gamma="auto"),{
    'C': [30],
    'kernel':['linear','rbf']
}, cv=5, return_train_score=False)
```

```python
cf.fit(iris.data,iris.target)
```

```python
cf.cv_results_
```
